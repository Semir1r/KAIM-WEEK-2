```python
import matplotlib.pyplot as plt

# Assume `cleaned_data` is your DataFrame
top_10_handset = cleaned_data["Handset Type"].value_counts().nlargest(10)

# Display the top 10 handsets
print("Top 10 Handsets Used by Customers:")
print(top_10_handset)

# Plotting the top 10 handsets
def plot_top_10_handsets(data, title="Top 10 Handsets Used by Customers"):
    plt.figure(figsize=(14, 7))
    data.plot(kind='bar', color='skyblue')
    plt.title(title)
    plt.xlabel("Handset Type")
    plt.ylabel("Number of Customers")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()

plot_top_10_handsets(top_10_handset)
```

```python
import seaborn as sns

# Identify the top 3 manufacturers
top_3_manufacturers = cleaned_data["Handset Manufacturer"].value_counts().nlargest(3)
print("Top 3 Handset Manufacturers:")
print(top_3_manufacturers)

custom_colors = ["#FF5733", "#33FF57", "#3357FF"] 
# Plot top 3 manufacturers
plt.figure(figsize=(8,5))
sns.barplot(x=top_3_manufacturers.values, y=top_3_manufacturers.index, palette="Blues_d")
plt.title('Top 3 Handset Manufacturers')
plt.xlabel('Number of Users')
plt.ylabel('Handset Manufacturer')
plt.show()
```

```python
# Create an empty dictionary to store the top 5 handsets per manufacturer
top_5_handsets_per_manufacturer = {}

# Iterate through the top 3 manufacturers
for manufacturer in top_3_manufacturers.index:
    # Filter the data for the current manufacturer
    filtered_data = cleaned_data[cleaned_data["Handset Manufacturer"] == manufacturer]
    
    # Get the top 5 handset types for this manufacturer
    top_5_handsets = filtered_data["Handset Type"].value_counts().nlargest(5)
    
    # Store the result in the dictionary
    top_5_handsets_per_manufacturer[manufacturer] = top_5_handsets

# Display the results
print("\nTop 5 Handsets per Top 3 Handset Manufacturers:")
for manufacturer, handsets in top_5_handsets_per_manufacturer.items():
    print(f"\nManufacturer: {manufacturer}")
    print(handsets)
```

```python
import matplotlib.pyplot as plt

# Assuming 'top_5_handsets_per_manufacturer' is already computed
# Initialize a figure for plotting
fig, axs = plt.subplots(len(top_5_handsets_per_manufacturer), 1, figsize=(10, 15), sharex=True)

# Plot data for each manufacturer
for i, (manufacturer, handsets) in enumerate(top_5_handsets_per_manufacturer.items()):
    axs[i].barh(handsets.index, handsets.values, color='skyblue')
    axs[i].set_title(f"Top 5 Handsets for {manufacturer}", fontsize=14)
    axs[i].set_xlabel("Count", fontsize=12)
    axs[i].set_ylabel("Handset Type", fontsize=12)
    axs[i].invert_yaxis()  # To display the highest count at the top

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Compute total session duration for each user
cleaned_data['Total_Duration'] = cleaned_data.groupby('MSISDN/Number')['Dur. (ms)'].transform('sum')

# Segment users into 5 deciles based on total duration
cleaned_data['Decile_Class'] = pd.qcut(cleaned_data['Total_Duration'], 5, labels=False)

# Compute total data (DL + UL) per decile class
cleaned_data['Total_Data'] = cleaned_data['Total DL (Bytes)'] + cleaned_data['Total UL (Bytes)']
decile_data = cleaned_data.groupby('Decile_Class')['Total_Data'].sum().reset_index()

# Display the results
print("Total Data per Decile Class:")
print(decile_data)
```

```python
# Correlation matrix for application data
app_data_columns = ['Social Media DL (Bytes)', 'Google DL (Bytes)', 'Youtube DL (Bytes)', 'Netflix DL (Bytes)', 'Gaming DL (Bytes)', 'Other DL (Bytes)']
correlation_matrix = cleaned_data[app_data_columns].corr()

# Display the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)

# Heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Application Data')
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Standardize the application data
scaler = StandardScaler()
app_data_scaled = scaler.fit_transform(cleaned_data[app_data_columns])

# Perform PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(app_data_scaled)

# Display explained variance ratio
print("Explained Variance by Component:", pca.explained_variance_ratio_)

# Plot the PCA result
plt.figure(figsize=(8, 6))
plt.scatter(pca_result[:, 0], pca_result[:, 1], alpha=0.7)
plt.title('PCA Result of Application Data')
plt.show()
```
